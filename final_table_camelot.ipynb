```python
import camelot

tabel = camelot.read_pdf('data/pdfs/PNPK 2023 GGK.pdf', pages='140')
```

```python
camelot.plot(tabel[0], kind='line').show()
```

```python
clean_tabel = tabel[0].df
clean_tabel
```

```python
import re
pattern = r'<th>\d+<\/th>|<th><\/th>'
tabel_html = clean_tabel.to_html().replace("\n", "")

tabel_html_clean = re.sub(pattern, '', tabel_html).replace("\n", "<br>")
print(tabel_html_clean)
```
