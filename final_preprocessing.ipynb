```python
def find_unique_characters(file_path):
    unique_characters = set()

    try:
        with open(file_path, 'r') as file:
            for line in file:
                # Remove whitespace and newline characters
                line = line.strip()
                # Add each character to the set
                for char in line:
                    if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
                        unique_characters.add(char)
    except FileNotFoundError:
        print("File not found.")
        return
    except Exception as e:
        print("An error occurred:", e)
        return

    return unique_characters
```

```python
def insert_newline(text, word_per_line):
    words = text.split()
    new_text = ""
    for i, word in enumerate(words, start=1):
        new_text += word + " "
        if i % word_per_line == 0:  # Add a newline every 25 words
            new_text += "\n"
    return new_text
```

```python
def remove_characters_from_file(input_file, output_file, chars_to_remove):
    try:
        # Open the input file for reading
        with open(input_file, 'r') as f_input:
            # Read the content of the file
            content = f_input.read()

            # Remove the specified characters from the content
            for char in chars_to_remove:
                content = content.replace(char, '')

        # Open the output file for writing
        with open(output_file, 'w') as f_output:
            # Write the modified content to the output file
            f_output.write(content)

        print(f"Characters {chars_to_remove} removed from '{input_file}'.")
        print(f"Modified content saved to '{output_file}'.")

    except FileNotFoundError:
        print("File not found!")
    except Exception as e:
        print(f"An error occurred: {e}")
```

# Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean_poster_final

```python
import fitz
from multi_column import column_boxes

text_list = []
filename = 'Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean.pdf'
filePath = f'./data\pdfs\\{filename}'
run = 0
# filename = 'Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean_poster.pdf'
# filePath = f'./data\pdfs-poster\\{filename}'
doc = fitz.open(filePath)
for page in doc:
    bboxes = column_boxes(page, header_margin=100, footer_margin=75, no_image_text=True)
    if run < 2:
        if run == 0 :
            bboxes.insert(0, bboxes.pop(3))
            run += 1
        elif run == 1:
            bboxes.insert(2, bboxes.pop(0))
            run += 1

    for rect in bboxes:
        ext_text = page.get_text(clip=rect, sort=True)
        text = ext_text.replace("", "")
        # text = ext_text.replace("\n", "")
        # text = text + "\n"
        print(text)
        # print("-" * 80)
        text_list.append(text)
```

```python
unique_characters = set()

for line in text_list:
    # Remove whitespace and newline characters
    line = line.strip()
    # Add each character to the set
    for char in line:
        if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
            unique_characters.add(char)

unique_chars = list(sorted(unique_characters))
for i, char in enumerate(unique_chars):
    print(i, char)
```

```python
import re

clean_text_list = []
chars = [unique_chars[0], unique_chars[11]]
pattern = r'[.:][ ]{0,}\d+(?:,\d+)+\s|[.:][ ]{0,}\d+-\d+\s|[.:][ ]{0,}\d+\s'

for line in text_list:
    ln = line.replace(chars[0], '')
    ln = ln.replace(chars[1], '')

    matches = re.findall(pattern, ln)  # Find all matches in the text
        
    if matches:
        temp = ln
        for match in matches:
            if "." in match:
                temp = temp.replace(match, ". ")
            elif ":" in match:
                temp = temp.replace(match, ": ")
        clean_text_list.append(temp)
    else:
        clean_text_list.append(ln)
```

```python
# A text file is created 
filename_clean = filename.replace(".pdf", "")
# filename_clean = "test"
file = open(f"./final_clean/{filename_clean}.txt", "w+")
file.write("")
file.close()

text = ""
for text in clean_text_list:
    file = open(f"./final_clean/{filename_clean}.txt", "a", encoding="utf-8")

    # # Appending the text into file
    file.write(text)
    file.write("\n")
    # print(text)
    # print(insert_newline(text, word_per_line = 10))
    # print("\n")
    
    # Close the file
    file.close()
```
