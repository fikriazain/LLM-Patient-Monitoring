# PNPK 2023 GGK

```python
from pypdf import PdfReader
```

```python
import re

filename = 'PNPK 2023 GGK_clean.pdf'
filePath = f'./data\pdfs\\{filename}'
loader = PdfReader(filePath)
pattern = r'- \d{1,3} -'
text = ""

for page in loader.pages:
    page_text = re.sub(pattern, "", page.extract_text().replace("jdih.kemkes.go.id", ""))
    text += page_text
```

```python
print(text)
```

```python
unique_characters = set()

for char in text:
    if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
        unique_characters.add(char)

unique_chars = list(sorted(unique_characters))
for i, char in enumerate(unique_chars):
    print(i, char)
```

```python
char = unique_chars[27]

text = text.replace(char, "→")
```

```python
# A text file is created 
filename_clean = filename.replace(".pdf", "")
file = open(f"./final/{filename_clean}.txt", "w+")
file.write("")
file.close()

file = open(f"./final/{filename_clean}.txt", "a", encoding="utf-8")

# # Appending the text into file
file.write(text)

# Close the file
file.close()
```

# nutrisi_pada_anak_dengan_penyakit_ginjal_clean_poster_final

### fitz

```python
import fitz

text_lst = []
filename = 'nutrisi_pada_anak_dengan_penyakit_ginjal - Copy_poster.pdf'
filePath = f'./data\pdfs-poster\\{filename}'
doc = fitz.open(filePath)
for page in doc:
    text = page.get_text()
    text_lst.append(text)
```

```python
text_list = []
for text in text_lst:
    text_list.append(text.replace("\n", " "))

for text in text_list:
    print(text)
```

```python
import re

pattern = r'[.]\d-\d\s|[.]\d\s|[.]\d(?:,\d+)+\s|\d(?:,\d){0,}:(?! \d{3})'

clean_text_list = []
for text in text_list:
    ln = text
    matches = re.findall(pattern, ln)  # Find all matches in the text
        
    if matches:
        temp = ln
        for match in matches:
            if "." in match:
                temp = temp.replace(match, ". ")
            elif ":" in match:
                temp = temp.replace(match, ": ")
        clean_text_list.append(temp)
    else:
        clean_text_list.append(ln)
```

```python
for clean_text in clean_text_list:
    print(clean_text)
```

```python
unique_characters = set()

for text in clean_text_list:
    # Remove whitespace and newline characters
    line = text.strip()
    # Add each character to the set
    for char in line:
        if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
            unique_characters.add(char)

unique_chars = list(sorted(unique_characters))
for i, char in enumerate(unique_chars):
    print(i, char)
```

```python
# A text file is created 
filename_clean = filename.replace(".pdf", "")
file = open(f"./final/{filename_clean}.txt", "w+")
file.write("")
file.close()

text = ""
for text in clean_text_list:
    file = open(f"./final/{filename_clean}.txt", "a")

    # # Appending the text into file
    file.write(text)
    file.write("\n")
    # print(text)

    # Close the file
    file.close()
```

# Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean_poster_final

```python
import fitz
from multi_column import column_boxes

text_list = []
filename = 'Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean.pdf'
filePath = f'./data\pdfs\\{filename}'
run = 0
# filename = 'Tatalaksana_Hemodialisis_pada_Anak_dan_Bayi_clean_poster.pdf'
# filePath = f'./data\pdfs-poster\\{filename}'
doc = fitz.open(filePath)
for page in doc:
    bboxes = column_boxes(page, header_margin=100, footer_margin=75, no_image_text=True)
    if run < 2:
        if run == 0 :
            bboxes.insert(0, bboxes.pop(3))
            run += 1
        elif run == 1:
            bboxes.insert(2, bboxes.pop(0))
            run += 1

    for rect in bboxes:
        ext_text = page.get_text(clip=rect, sort=True)
        # text = ext_text.replace("", "")
        text = ext_text.replace("\n", " ")
        # text += "\n"
        # print("-")
        print(text)
        # print("-")
        text_list.append(text)
```

```python
for line in text_list:
    # Remove whitespace and newline characters
    print(line)
```

```python
unique_characters = set()

for line in text_list:
    # Remove whitespace and newline characters
    line = line.strip()
    # Add each character to the set
    for char in line:
        if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
            unique_characters.add(char)

unique_chars = list(sorted(unique_characters))
for i, char in enumerate(unique_chars):
    print(i, char)
```

```python
# A text file is created 
filename_clean = filename.replace(".pdf", "")
# filename_clean = "test"
file = open(f"./final/{filename_clean}.txt", "w+")
file.write("")
file.close()

text = ""
for text in clean_text_list:
    file = open(f"./final/{filename_clean}.txt", "a", encoding="utf-8")

    # # Appending the text into file
    file.write(text)
    file.write("\n")
    # print(text)
    # print(insert_newline(text, word_per_line = 10))
    # print("\n")
    
    # Close the file
    file.close()
```

# Tatalaksana_Penyakit_Ginjal_Kronik_pada_Anak_clean

```python
from langchain_community.document_loaders import PyPDFLoader

filename = 'Tatalaksana_Penyakit_Ginjal_Kronik_pada_Anak_clean_split_final.pdf'
filePath = f'./data\pdfs-final\\{filename}'
loader = PyPDFLoader(filePath)
pages = loader.load_and_split()
```

```python
text_list = []
for page in pages:
    text_list.append(page.page_content)
```

```python
for text in text_list:
    print(text)
```

```python
unique_characters = set()

for text in text_list:
    # Remove whitespace and newline characters
    line = text.strip()
    # Add each character to the set
    for char in line:
        if not char.isalnum() and not char.isspace():  # Check if the character is not a letter
            unique_characters.add(char)

unique_chars = list(sorted(unique_characters))
for i, char in enumerate(unique_chars):
    print(i, char)
```

```python
clean_text_list = []
for text in text_list:
    if unique_chars[0] in text:
        clean_text_list.append(text.replace(unique_chars[0], ""))
    else:
        clean_text_list.append(text)
```

```python
for text in clean_text_list:
    print(text)
```

```python
import re

pattern = r'[.]\s{0,}\d\s|[.]\s{1,}\d+(?:,\d+){0,}|:\s{0,}\d+(?:,\d+){0,}'

clean_text_list2 = []
for text in clean_text_list:
    ln = text
    matches = re.findall(pattern, ln)  # Find all matches in the text
        
    if matches:
        temp = ln
        for match in matches:
            if "." in match:
                temp = temp.replace(match, ". ")
            elif ":" in match:
                temp = temp.replace(match, ": ")
        clean_text_list2.append(temp)
    else:
        clean_text_list2.append(ln)
```

```python
# A text file is created 
filename_clean = filename.replace(".pdf", "")
file = open(f"./final/{filename_clean}.txt", "w+")
file.write("")
file.close()

with open(f"./final/{filename_clean}.txt", "a", encoding="utf-8") as file:
    for text in clean_text_list2:
        file.write(text)
        file.write("\n")
```
